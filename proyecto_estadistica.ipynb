```
import pandas as pd
df=pd.read_csv('casa.csv')
import statsmodels.formula.api as smf
```

```
# -*- coding: utf-8 -*-

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

sns.set_style('darkgrid')

datos=pd.read_csv('casa.csv')
nuevo=datos[['age','distance','stores']]
regression=LinearRegression()
regression.fit(nuevo,datos['area']) #train
#predict
data_predict=[[32,84.87882,10]]
y_hat=regression.predict(data_predict)
print(y_hat,"-------------prediccion del precio--------------")
print("intercepto o b_0 ",regression.intercept_)
print("coeficientes", regression.coef_)
sns.regplot(x='age',y='area',data=datos,line_kws={'color':'red'})

#individual simple linear regression

```

```
counts, start, dx, _ = scipy.stats.cumfreq(datos['age'],numbins=415)
x = np.arange(counts.size) * dx + start

plt.plot(x, counts, 'ro')
plt.xlabel('age')
plt.ylabel('Cumulative Frequency')

plt.show()
```
